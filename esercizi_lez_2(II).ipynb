```python
#esercizio 5
stipendi=np.array([50000, 105250, 55000, 89000])
costo_tot=sum(stipendi)
print("all'inizio: ", costo_tot)
#----
stipendi[1]=stipendi[1]*1.15
print(stipendi)
#in questo modo arrotonda a int...
#----
stipendi=np.array([50000, 105250, 55000, 89000], dtype=float)
stipendi[1]=stipendi[1]*1.15
print(stipendi)
#----
stipendi[0]=stipendi[0]*1.2
print(stipendi)
#----
stipendi[2]=stipendi[2]*1.10
stipendi[3]=stipendi[3]*1.10
print(stipendi)
#----
costo_tot_f=sum(stipendi)
print("dopo gli aumenti: ", costo_tot_f)
aumento=costo_tot_f-costo_tot
print("l'aumento è estato di ", aumento, " euro per l'azienda")
```
