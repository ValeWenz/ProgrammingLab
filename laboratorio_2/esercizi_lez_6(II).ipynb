```python
#esercizio 1
sales = pd.DataFrame(
    data={
        "employee": [
            "Katrina",
            "Guanyu",
            "Jan",
            "Roman",
            "Jacqueline",
            "Paola",
            "Esperanza",
            "Alaina",
            "Egweyn",
        ],
        "sales": [14, 17, 6, 12, 8, 3, 7, 15, 5],
        "year": [2018, 2019, 2020, 2018, 2020, 2019, 2019, 2020, 2020],
    }
)
sales

#1 vendite maggiori di 10   sales[sales["sales"]>10]
sales.query("sales>10")
#2 dati del 2018   sales[sales["year"]==2018]
sales.query("year==2018")
#3 vendite maggiori di 13 e anno 2018
sales.query("sales>13 and year==2018")
#4 tutto TRANNE i casi in cui le vendite sono maggiori di 13 e anno==2018
sales.query("not sales>13 and not year==2018")
#5 vendite divise per 3 sono maggiori di 3
sales.query("sales//3 >3")
#6 dipendenti i cui nomi sono dopo la J
sales.query('employee>"J"')
```

```python
#esercizio 2

# Caricamento del dataset
url = 'https://zenodo.org/record/5898311/files/vgsales.csv'
df = pd.read_csv(url)
df.head()


#1 Quanti videogiochi sono stati pubblicati
df.shape[0]

#2 Quali sono i generi più popolari? Mostralo con un bar plot
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="Genre")
plt.title("popolarità dei generi")
plt.xlabel("genere")
plt.xticks(rotation=45)
plt.ylabel("conteggio")
plt.show()

#3 Plotta l'evoluzione del numero di giochi pubblicati nel tempo
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="Year")
plt.title("evoluzioni di pubblicazioni nel tempo")
plt.xlabel("anno")
plt.xticks(rotation=45)
plt.ylabel("conteggio")
plt.show()

#4 Prova a riprodurre questo plot
plt.figure(figsize=(10,6))
df_aux=df["Genre"].value_counts().nlargest(5).index
aiuto=[g in df_aux for g in df["Genre"]]
df_aux2=df[aiuto]
df_grafico=df_aux2.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
df_grafico.plot(kind='bar', stacked=True)
plt.xlabel("genre")
plt.ylabel("milioni di unità")
plt.title("vendite per regione per genere")
plt.show()


```
